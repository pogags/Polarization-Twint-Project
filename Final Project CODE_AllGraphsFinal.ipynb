```python
#Health Care
df16 = pd.read_csv("healthcareTweets2016.csv")
df16.insert(3, "Year", 2016)
df17 = pd.read_csv("healthcareTweets2017.csv")
df17.insert(3, "Year", 2017)
df18 = pd.read_csv("healthcareTweets2018.csv")
df18.insert(3, "Year", 2018)
df19 = pd.read_csv("healthcareTweets2019.csv")
df19.insert(3, "Year", 2019)
df20 = pd.read_csv("healthcareTweets2020.csv")
df20.insert(3, "Year", 2020)
healthCareDF = pd.concat([df16, df17, df18, df19, df20], axis=0)
#healthCareDF

resList = []

for t in healthCareDF['tweet'] :
        t = t.replace('\n','.')
        tweetIndex = t.find('https://t.co')
        if tweetIndex != -1 :
            alteredTweet = t[:tweetIndex]
        else :
            alteredTweet = t
  
        newTweet = " ".join(filter(lambda x:x[0]!='@', alteredTweet.split()))
        resList.append(newTweet)    
                       
healthCareDF['cleanTweet'] = resList

#healthCareDF
```

```python
def decontract(phrase):
    phrase = re.sub(r"won\'t", "will not", phrase)
    phrase = re.sub(r"can\'t", "can not", phrase)
    phrase = re.sub(r"don\'t", "do not", phrase)
    phrase = re.sub(r"n\'t", " not", phrase)
    phrase = re.sub(r"\'re", " are", phrase)
    phrase = re.sub(r"\'s", " is", phrase)
    phrase = re.sub(r"\'d", " would", phrase)
    phrase = re.sub(r"\'ll", " will", phrase)
    phrase = re.sub(r"\'t", " not", phrase)
    phrase = re.sub(r"\'ve", " have", phrase)
    phrase = re.sub(r"\'m", " am", phrase)
    phrase = re.sub(r"don't","do not", phrase)
    return phrase

lemmatizer = WordNetLemmatizer()
```

```python
#Immigration all years

fString = []
for twt in immigrationDF['cleanTweet'] :
    fString.append(twt)
    
fString = "".join(fString)
fString = decontract(fString)
fString = decontract(fString).replace('...',' ')
stop_words = set(stopwords.words('english'))
word_tokens = word_tokenize(fString)
filtered_sentence = [w for w in word_tokens if not w in stop_words]
for wrd in range(len(filtered_sentence)) :
    filtered_sentence[wrd] = lemmatizer.lemmatize(filtered_sentence[wrd])
filtered_sentence = " ".join(filtered_sentence)
filtered_sentence

wordcloud = WordCloud().generate(filtered_sentence)

plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

plt.show()  
```
